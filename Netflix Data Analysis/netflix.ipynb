## Correct Data Type
### Here date_added should be date type instead of object

```python
# Strip leading/trailing space
df['date_added'] = df['date_added'].str.strip()

# Convert 'date_added' from object to datetime with specified format and handle errors
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce') # Verify the type conversion
df.dtypes
```
